## Heatmap

```python
data_ciclista=data[((data['incidente_c4']=="cadáver-accidente automovilístico") | (data['incidente_c4']=="cadáver-atropellado"))&(data['año_cierre']==2018)]
```

```python
data_ciclista['latitud'] = data_ciclista['latitud'].astype(float)
data_ciclista['longitud'] = data_ciclista['longitud'].astype(float)
```

```python
data_ciclista.head()
```

```python
import folium
import csv
from folium.plugins import HeatMap

heatmap_map = folium.Map(location=[19.4326, -99.12766], zoom_start=2)

heat_df = data_ciclista.dropna(axis=0, subset=['latitud','longitud'])
heat_data = [[row['latitud'],row['longitud']] for index, row in heat_df.iterrows()]

HeatMap(heat_data).add_to(heatmap_map)
```
